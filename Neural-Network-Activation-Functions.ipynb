```python
import numpy as np
```

### ReLU activation on hidden layer and sigmoid activation on output layer

#### CREATING ReLU ACTIVATION FUNCTION
    f(x) = 0 for x<0
         = x for x>0

```python
def relu(x):
    return (max(0,x))
```

#### CREATING Sigmoid ACTIVATION FUNCTION
    f(x) = 1 / (1+e^-x)

```python
def sigmoid(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
data=np.array([2,3])
x0 = data[0]
x1 = data[1]
```

#### INPUT LAYER

```python
input_layer =np.array([x0,x1])
input_layer
```

```python
weights = {
    'x2':np.array([1,1]),
    'x3':np.array([-1,1]),
    'x4':np.array([-2,1]),
    'x5':np.array([1,0]),
    'x6':np.array([1,-1])
}
```

#### HIDDEN LAYER 1

```python
x2 = (input_layer*weights['x2']).sum()
x3 = (input_layer*weights['x3']).sum()
```

```python
x2 = relu(x2)
x3 = relu(x3)
```

```python
hidden_layer1 = np.array([x2,x3])
hidden_layer1
```

#### HIDDEN LAYER 2

```python
x4 = (hidden_layer1*weights['x4']).sum() 
x5 = (hidden_layer1*weights['x5']).sum()
```

```python
x4 = relu(x4)
x5 = relu(x5)
```

```python
hidden_layer2 = np.array([x4,x5])
hidden_layer2
```

#### OUTPUT LAYER

```python
x6 = (hidden_layer2*weights['x6']).sum()
x6 = sigmoid(x6)
```

```python
output = np.array([x6])
output
```
