# Create a single layered neural network with 2 input values

```python
import numpy as np
```

### INPUT LAYER
    - Two input : [2,3]
    - Therefore, two nodes are x0 and x1

```python
data = np.array([2,3])
```

```python
x0 = data[0]
x1 = data[1]
```

```python
input_layer = np.array([x0,x1])
input_layer
```

### 2 HIDDEN LAYERS:
    hidden layer 1 : 
        --> nodes : [x2,x3]
        --> weights : x2 = [1,1]
                      x3 = [-1,1]
    hidden layer 2 : 
        --> nodes : [x4,x5]
        --> weights : x4 = [-2,1]
                      x5 = [1,0]
                      

```python
weights = {
    'x2':np.array([1,1]),
    'x3':np.array([-1,1]),
    'x4':np.array([-2,1]),
    'x5':np.array([1,0]),
    'x6':np.array([1,-1])
}
```

```python
#finding weighted sum of inputs at HIDDEN LAYER 1
# hidden layer has x2,x3
# so calculating WEIGHTED SUM at x2 and x3

x2 = (input_layer*weights['x2']).sum()
x3 = (input_layer*weights['x3']).sum()
print(x2,x3)
```

```python
hidden_layer_1 = np.array([x2,x3])
hidden_layer_1
```

```python
#finding weighted sum of inputs at HIDDEN LAYER 2
# hidden layer has x4,x5
# so calculating WEIGHTED SUM at x4 and x5

x4 = (hidden_layer_1*weights['x4']).sum()
x5 = (hidden_layer_1*weights['x5']).sum()
print(x4,x5)
```

```python
hidden_layer_2 = np.array([x4,x5])
hidden_layer_2
```

#### OUTPUT LAYER
    - This layer has node 'x6'
    - It has two connections i.e. with x4 and x5 from HIDDEN LAYER 2
    - So it will have two weights (already defined in dict) : [1,-1]

```python
x6 = (hidden_layer_2*weights['x6']).sum()
x6
```

```python
output_layer = np.array([x6])
output_layer
```

### CONCLUSION
    - Data received at output layer is -14
    - So it means tht if we give 2,3 as input then we get output as -14
